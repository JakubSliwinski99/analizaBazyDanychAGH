#### Zad 7
Znajdź wszystkich klientów z Polski, Nigerii lub Bangladeszu.

```python
result_set = db.execute("SELECT first_name, last_name FROM customer " + 
                        "LEFT JOIN address ON address.address_id = customer.address_id " + 
                        "LEFT JOIN city ON city.city_id = address.city_id " + 
                        "LEFT JOIN country ON country.country_id = city.country_id " + 
                        "WHERE country.country = 'Poland' OR country.country = 'Nigeria' OR country.country = 'Bangladesh' ")  
for r in result_set:  
    print(r)
```
